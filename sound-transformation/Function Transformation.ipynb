First, we import an audio file in the form of a function which we call $f(x)$.

Let's graph the function $f(x)$:

```python
x = np.arange(0, x_max, 1/sr).astype(np.float32) # x-axis values

plt.figure(figsize=(14,5))
plt.ylabel('amplitude')
plt.xlabel('time (sec)')
ax = plt.gca()
ax.set_ylim([-1, 1])
ax.set_xlim([0, 18])

plt.plot(x,f(x))
```

Next we perform a transofrmation on the function $f(x)$. Let's define $y=f(x-1)$ and then plot the original function $f(x)$ together with its transformation:

```python
y=f(x)+f(x-.3)


plt.figure(figsize=(14,6))

plt.subplot(211)
ax = plt.gca()
ax.set_ylim([-1, 1])
ax.set_xlim([0, 18])
plt.plot(x, f(x), 'b')
# naming the y axis
plt.ylabel('amplitude, orig')

plt.subplot(212)
ax = plt.gca()
ax.set_ylim([-1, 1])
ax.set_xlim([0, 18])
plt.plot(x, y, 'm')
 
# naming the x axis
plt.xlabel('time (sec)')
# naming the y axis
plt.ylabel('amplitude, new')
```
